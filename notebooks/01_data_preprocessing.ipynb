# Data Preprocessing for Bengali Empathetic LLM

This notebook preprocesses the raw Bengali conversation data for fine-tuning an empathetic language model.

## Cell 1 – Load and Inspect Raw JSON

```python
import json
import pandas as pd
from sklearn.model_selection import train_test_split
from pathlib import Path

raw_path = Path("../data/raw/bengali_conversations.json")
with raw_path.open("r", encoding="utf-8") as f:
    conversations = json.load(f)

print("Total examples:", len(conversations))
conversations[:3]
```

## Cell 2 – Convert to Instruction Format

```python
formatted = []
for conv in conversations:
    user = conv["user"]
    assistant = conv["assistant"]
    text = f"""Below is a conversation between a user and an empathetic assistant.

### User:
{user}

### Assistant:
{assistant}
"""
    formatted.append({"text": text})

df = pd.DataFrame(formatted)
print(df.head())
print("Total formatted examples:", len(df))
```

## Cell 3 – Split into Train / Val / Test

```python
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print("Train size:", len(train_df))
print("Val size:", len(val_df))
print("Test size:", len(test_df))
```

## Cell 4 – Save Splits

```python
out_dir = Path("../data/processed")
out_dir.mkdir(parents=True, exist_ok=True)

train_df.to_json(out_dir / "train.json", orient="records", force_ascii=False, indent=2)
val_df.to_json(out_dir / "val.json", orient="records", force_ascii=False, indent=2)
test_df.to_json(out_dir / "test.json", orient="records", force_ascii=False, indent=2)

print("Saved:", list(out_dir.iterdir()))
```
